## 2.2 Problem 2

### (a)

```r
install.packages('ISLR')
```

```r
library('ISLR')
```

```r
data('Auto')
```

```r
lm.fit = lm(mpg~horsepower, data = Auto)
summary(lm.fit)
```

#### i.

```r
f <- summary(lm.fit)$fstatistic
p <- pf(f[1],f[2],f[3],lower.tail=F)
p
```

Since the null hypothesis is that there is no relationship between mpg and horsepower while the p-value corresponding to the F-statistic is 7.03198902940446e-81, it can be concluded that the null hypothesis is false and there is a relationship between mpg and horsepower.

#### ii.

```r
paste("Varriability: ",summary(lm.fit)$r.squared * 100, " %")
RSE <- summary(lm.fit)$sigma
mean <- getElement(summary(Auto$mpg), 'Mean')
paste("Error: ",RSE/mean * 100, "%")
```

With 60.5948257889435 % varriability and 20.9237140669148 % error, horsepower can explain mpg

#### iii.

```r
paste("Coefficient of horsepower: ",summary(lm.fit)$coefficients[2])
```

The relationship is negative since the coefficient of horsepower is -0.157844733353654  (negative).

#### iv.

```r
print("Confidence Interval")
predict(lm.fit, data.frame(horsepower = 98), interval = "confidence")
print("Prediction Interval")
predict(lm.fit, data.frame(horsepower = 98), interval = "prediction")
```

Confidence Interval: [23.97308 - 24.96108]

Prediction value: 24.46708 (Same for both cases)

### (b)

```r
plot(Auto$horsepower, Auto$mpg, main = "Least squares regression")
abline(lm.fit, lwd = 3, col = 'red')
```

### (c)

```r
par(mfrow = c(2,2))
plot(lm.fit, lwd = 3, col = 'blue') 
```

Something that stands out is the residuals vs leverage graph. There are many outliers with regard to Standardized Residuals but an even spread which might say something about the data; there is one big outlier for leverage that might be just an outlier. There are curves in Residual vs Fitted graph which indicates a less linear fit
